```python
import ast
import pandas as pd
# df = pd.read_csv("ocrx_dailymed.csv")
def parse_df(df):
    columns_to_convert = ['form','roa','active_ingredient']
    def parse(el):
        if type(el) == str:
            return ast.literal_eval(el)
        return []
    for column in columns_to_convert:
        df[column] = df[column].apply(parse)
    return df

df = parse_df(pd.read_csv("ocrx_dailymed.csv"))
```
