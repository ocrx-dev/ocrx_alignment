```python
import pandas as pd
new_match_infos = [
    ("ingrédients actifs", "new_match_active_ingredient.csv"),
    ("formes", "new_match_form.csv"),
    ("ROA", "new_match_roa.csv")

]
all_dfs = []
for match_name, match_fn in new_match_infos:
    match_df = pd.read_csv(f'{match_fn}').dropna().drop_duplicates(subset=['code_dailymed','code_ocrx'])
    match_df = match_df.astype(str)
    all_dfs.append(match_df)
```

```python
df = all_dfs[0]
```

```python
pd.DataFrame([(key, df['label'].iloc[0], ' | '.join(df['code_dailymed'].values))for key, df in df.groupby("code_ocrx")])
```
