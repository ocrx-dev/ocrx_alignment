```python
def check_df(df):
    ft_ids = set(df[df['composant_type'] == 'FT']['num_linked'].values)
    sa_ids = set(df[df['composant_type'] == 'SA']['num_linked'].values)
    assert sa_ids.union(ft_ids) == sa_ids
```

```python
def find_mapping_dfs(code_cis):
    def to_str_dose(row):
        if str(row['dose']) == row['dose'] and str(row['dose_ref']) == row['dose_ref']:
            return ' / '.join([row['dose'],row['dose_ref']])
        elif str(row['dose']) == row['dose']:
            return row['dose']
        else:
            return ''
    sub_df_raw = code_cis_df.get(code_cis,None)
    if sub_df_raw is None:
        return []
    sub_df = sub_df_raw[sub_df_raw['composant_type'] == 'SA'].sort_values('num_linked')
    compos = [(row['denomination'], to_str_dose(row)) for _, row in sub_df.iterrows()]
    return compos
```

```python
def to_df(a_set,term):
    sorted_set = list(sorted(a_set))
    df = pd.DataFrame()
    df['ansm_code'] = [i for i in range(len(sorted_set))]
    df['ansm_label'] = [el.upper() for el in sorted_set]
    df.to_csv(f'export/ansm_{term}_fr.csv',index=False)
def construct_drug_name(row):
    form, roa = row['form'], row['roa']
    compo_str = ' | '.join([f'{el[0]} AVEC DOSAGE DE {el[1]}' for el in row['compos']])
    if form is not None:
        compo_str += f' SOUS FORME DE {form}'
    if roa is not None:
        roa_splitted = ' '.join(roa.split(';'))
        compo_str += f' PAR VOIE {roa_splitted}'
    return compo_str.upper()
def export_drug(df_bdpm):
    unique_compos = {f'{compo[0]} AVEC DOSAGE DE {compo[1]}'.upper() for el in df_bdpm['compos'].values for compo in el}
    unique_ing = {compo[0].upper() for el in df_bdpm['compos'].values for compo in el}
    unique_str = {compo[1].upper() for el in df_bdpm['compos'].values for compo in el}
    to_df(unique_compos,'component')
    to_df(unique_ing,'active_ingredient')
    to_df(unique_str,'strength')
    df_bdpm['drug_name'] = [construct_drug_name(row) for _, row in df_bdpm.iterrows()]
    new_df = df_bdpm[['code_cis','drug_name']].drop_duplicates().copy()
    new_df.columns = ['ansm_code','ansm_label']
    new_df.to_csv("export/ansm_drug_fr.csv",index=False)
#     return compos

export_drug(df_bdpm)
```
